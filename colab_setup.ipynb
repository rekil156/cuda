### BAsic example

```
%%cu
#include <stdio.h>
#include <stdlib.h>
__global__ void add(int *a, int *b, int *c) {
*c = *a + *b;
}
int main() {
int a, b, c;
// host copies of variables a, b & c
int *d_a, *d_b, *d_c;
// device copies of variables a, b & c
int size = sizeof(int);
// Allocate space for device copies of a, b, c
cudaMalloc((void **)&d_a, size);
cudaMalloc((void **)&d_b, size);
cudaMalloc((void **)&d_c, size);
// Setup input values  
c = 0;
a = 3;
b = 5;
// Copy inputs to device
cudaMemcpy(d_a, &a, size, cudaMemcpyHostToDevice);
  cudaMemcpy(d_b, &b, size, cudaMemcpyHostToDevice);
// Launch add() kernel on GPU
add<<<1,1>>>(d_a, d_b, d_c);
// Copy result back to host
cudaError err = cudaMemcpy(&c, d_c, size, cudaMemcpyDeviceToHost);
  if(err!=cudaSuccess) {
      printf("CUDA error copying to Host: %s\n", cudaGetErrorString(err));
  }
printf("result is %d\n",c);
// Cleanup
cudaFree(d_a);
cudaFree(d_b);
cudaFree(d_c);
return 0;
}
```

#### Pure Python vector add

```
import numpy as np
from timeit import default_timer as time
```

```
def vectorADD(A,B,C):
  for i in range(A.size):
    C[i] = A[i] +B[i]
```

```
N = 10000000
A = np.ones(N,dtype=np.float32)
B = np.ones(N, np.float32)
C = np.zeros(N, np.float32)

start = time()
vectorADD(A,B,C)
end = time() - start 
print("Time taken -%f" %end)
print("First 5 in C -",C[:5])
```

@vectorize - 
* target function should be scalar ie input and output should be scalars
* can accept any number of scalar inputs but returns a single scalar oputput
* signature: string - input and output data type -> `@vectorize(['output_signature('input_signature1', ....)]`
